```python
import numpy as np
import pickle
```

```python
with open("results/stats.pkl", "rb") as f:
    stats = pickle.load(f)
```

```python
stats[0].keys()
```

```python
all_srs = {
    s:
        {
            k: stats[s]["incremental_success_rates"][k][-1]
            for k in stats[s]["incremental_success_rates"].keys()
        }
    for s in range(5)
}
all_srs_mean_std = {
    k: (
        np.round(
            np.mean([all_srs[s][k] for s in all_srs.keys()]), 2
        ),
        np.round(
            np.std([all_srs[s][k] for s in all_srs.keys()]), 2
        ),
    )
    for k in all_srs[0].keys()
}
"inc", all_srs_mean_std
```
